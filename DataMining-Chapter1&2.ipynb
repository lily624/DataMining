## 数据的对象与属性类型

数据集由数据对象组成，一个数据对象代表一个实体。比如，在制造业的生产数据中，对象是产品；在医疗数据库中，对象可以是患者。数据对象用属性来描述。数据对象又称为样本、实例、数据点或者对象。如果对象存放在数据库中，他们就是数据元组。直观的说，数据库中的行代表数据对象，列代表数据属性。

### 属性

属性就是一个数据字段，用来表示数据对象的一个特征。很多时候，属性、维、特征和变量是可以相互使用的。数据挖掘和数据库的专业人士一般用“属性”。比如描述一个西瓜的属性可能包括西瓜纹路、重量、敲击西瓜的声音、西瓜皮的质感等等。给定属性的观测值叫**观测**。用来描述一个给定对象的一组属性就叫**属性向量**（**或特征向量**），比如 西瓜 = \[纹路、重量、敲击西瓜的声音\]。

涉及一个属性的数据分布就叫做单变量的，2个属性的数据分布就叫双变量。

**标称属性**

    标称属性（nomial attribute）的值是一些符号或者事物的名称，每个值用来代表某种类别、编码或者状态，所以标称属性被看做是分类。比如西瓜纹路和敲击西瓜的声音是描述西瓜的属性，西瓜的纹路可能是“清晰”、“稍糊”和“模糊”，敲击西瓜的声音的可能值是：“浊响”、“沉闷”和“清脆”。西瓜的纹路 和 敲击西瓜的声音就是标称属性。
    
    在标称属性上，数学运算是没有意义的。标称属性是那些不打算定量地使用的那些属性。
    
**二元属性**

    二元属性是一种标称属性，只有两种类别或者状态：0或1.通常0表示该属性不出现，1表示出现。比如某个被测者描述患者是否抽烟，1-表示抽烟，0表示不抽烟。类似地，核酸检测结果具有两种可能性，其中1表示患者阳性，0表示患者阴性。
    
    二元属性又有两种可能性：对称和非对称。对称的属性指的是，关于哪个结果应该用0或者1，编码没有偏好，两种状态具有相同的权重。非对称属性就是，状态的结果不是同样重要的，有不一样的权重。比如HIV化验结果阴阳性，1对最重要的结果编码（阳性），0就是阴性的结果编码。
    
**序数属性**

    序数属性的可能只之间具有有意义的序或秩评定（ranking），但是相继之间的差是未知的。比如西瓜一个属性“甜的程度”，具有3个可能值---不甜、一般、很甜。这些值具有有意义的先后次序（对应于递增的含糖度），但是我们不能说很甜比一般会甜多少。类似的例子还有乘机（A+,A-,B+,B-），职位（教授、副教授、讲师等），客户满意度（0-很不满意，1-不太满意，2-中性，3-满意，4-非常满意）
    序数属性的中心趋势可以用众数、中位数表示，但不能用定义均值。
    
**标称属性、二元属性、序数属性都是定性的，没有实际大小或者数量，是类别的计算机编码。**


**数值属性**

    数值属性是定量的，可度量的量，用整数或者数值表示。数值属性可以是区间标度，也可以是比率标度。
    区间标度： 用相等的单位尺度度量，区间属性的值是有序的，可以是正、0或负。也允许比较和定量评估值之间的差。比如：温度，假设有很多天的室外温度值，其中每天是一个对象。可以对这些值排毒，就可以得到关于温度的秩评定，也可以量化不同值之间的差，比如20℃比5℃高出15℃。这些属性没有真正的零点或者说没有绝对的零点，0℃不表示“没有温度”。但是不能用比率谈论这些值，比如我们不能说10℃比5℃温暖2倍。
    
    比率标度：具有固有零点的数值属性，也就是说，可以用一个值是另一个值的倍数说比率来解释比率标度。比率标度属性，比如工作年限（对象是雇员）和字数（对象是文档）等计数属性，重量、高度、速度和货币量。
    
**离散属性和连续属性**

    可以有很多种方法来组织属性类型，这些类型是不互斥的。
    机器学习领域开发的分类算法通常把属性分为离散的和连续的。
    离散属性：具有有限或无限可数个数，因此是离散的。
    连续属性：不是离散的就是连续的，实践中连续属性一般用浮点变量表示。


## 数据的基本统计描述

基本统计藐视可以用来识别数据的性质，凸显哪些数据值应该视为噪声或者离群点。

### 中心趋势度量：均值、中位数和众数

数据的中心趋势度量数据分布的中心位置，就是给定一个属性，它的值大部分落在哪里？通常是均值、中位数、众数和中列数。

比如某个属性X，已经对一个数据对象集记录了他们的值，也就是$x_{1},x_{2},...,x_{N}$.

**均值**

数据集“中心”，最常用的就是（算数）平均值， 
$$\bar{X} = \frac{\sum_{i=1}^{N} x_{i}}{N}$$

对于每一值$x_{i}$都有一个相关联的$w_{i}$的情况，计算均值就变成了加权平均值，也就是

$$\bar{X} = \frac{\sum_{i=1}^{N} w_{i} x_{i}}{\sum_{i=1}^{N} w_{i}}$$

**中位数**

    按递增排序，排最中间的那个数就是中位数，有偶数个观测只是，取最中间两个值的平均值。

**众数**

    集中出现最频繁的值，具有一个众数的数据集成为单峰（unimodal)的,具有两个或以上的数据集是多峰（multimodal）的。

**中列数（midrange）**

    中列数是数据集的最大和最小值的平均值。


### 度量数据散布：极差、四分位数、方差、标准差和四分位数极差

数据的散布是指数据如何分散的，常见的度量是数据的极差、四分位数、四分位数极差、五数概括和盒图，以及数据的方差和标准差。


**极差、四分位数和四分位数极差**

    极差：最大值和最小值之差
    分位数：取自数据分布的每隔一定间隔上的点，把数据分成基本上大小相等的连贯集合。
         第k个q-分位数是值x,使得小于x的数据值最多为k/q,而大于x的数据值最多为（q-k)/q,其中k是整数。
         
    四分位数：四分位数是3个数据点，把数据划分成4个相等的部分，每部分表示数据分布的四分之一
    四分位数极差（IQR）：第1个和第3个四分位数之间的距离是散步的一种简单度量，是被数据的中间一般所覆盖的范围。这个距离就是四分位数极差。

**五数概况、盒图与离群点**
    
    五数概括：中位数（Q2)，四分位数Q1和Q3、最小和最大观测值组成，按次序Minimun, Q2, Median, Q3, Maximum.
    盒图（BoxPlot)：盒的端点一般是四分位数，使得盒的长度是四分位数极差IQR；中位数用盒内的线标记，盒外的两条胡须延伸到最小和最大观测值。
    
**标准差和方差**

    标准差和方差都是指出数据分布的散布程度。标准差越小，说明观测趋向于靠近均值。标准差是方差的平方根。
    标准差度量关于均值的发散，仅当选择均值作为中心度量的时候使用。当所有的观测值都具有相同值时，标准差为0.否则，标准差大于0.

## 度量数据的相似性和相异性

**簇**

    是数据对象的几何，使得同一个簇中的对象互相相似，而与其他簇中的对象相异。

相似性和相异性都称为临近性(proximity)。

**数据矩阵（data matrix)**

数据矩阵也称为对象-属性结构，这种数据结构用n x p矩阵存放n个数据对象。

**相异性矩阵（dissimilarity matrix)**

d(i,j)是对象i和j之间的相异性或“差别”的度量，一般来说是非负的。sim(i,j)是对象i和j之间的相似性。sim(i,j) = 1 - d(i,j).许多聚类和最近邻算法都在相异性矩阵上运行。

### 邻近性度量

1. 标称属性的邻近性度量

  假设一个标称属性的状态数目是M。两个对象i和j之间的相异性可以根据不匹配率来算：
  
  $$d(i,j) = \frac{p-m}{p} $$
  
  其中，m是匹配数目，p是刻画对象的属性总数。
  
2. 二元属性的邻近性度量

  如果所有的二元都被看做具有相同的权重，则用过一个两行两列的列联表。对称的二元相异性：
  
  $$d(i,j) = \frac{r+s}{q+r+s+t}$$
  
  其中，q是对象i和j都取1的属性数，r是对象i中取1，对象j中取0的属性数，s是对象i中取0，对象j中取1的属性数，t是是对象i，j中都取0的属性数。
  
  对于非堆成的二元属性，两个状态不是同等重要的,非对称的二元相异性：
  $$d(i,j) = \frac{r+s}{q+r+s}$$
  
  对于非对称的二元相似性，计算公式如下：
  $$sim(i,j) = 1 - d(i,j) = \frac{q}{q+r+s}$$
  
  sim(i,j)被称作Jaccard系数。
